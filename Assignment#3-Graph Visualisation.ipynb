**Import libraries**

```python
#Load needed packages
%matplotlib inline
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

**Read data into notebook**

```python
data =pd.read_csv("https://github.com/latoyagirl/DATA620-Web-Analytics/blob/master/soc-sign-bitcoinalpha%20copy.csv")
```

The data set was taken from the following website: https://snap.stanford.edu/data/soc-sign-bitcoin-alpha.html. It contains 24187 records.

```python
data.head()
```

```python
len(data.Source.unique()) #   number of unique raters
```

```python
len(data.Target.unique())#number of unique ratees
```

**Create Graph**

```python
G=nx.from_pandas_edgelist(data, 'Source', 'Target', create_using=nx.DiGraph() )
```

```python
nx.draw(G, with_labels=True, node_size=150, alpha=0.3, arrows=True)
plt.show()
```

**Basic Analysis on Graph**

```python
nx.info(G) # number of nodes and edges
```

```python
nx.diameter(G)
```

The diameter of a network graph is the greatest path between two nodes.A diameter measure cannot be taken  because the nodes are not all connected to each other, thus creating an infinite path, which is a meaningless value.
