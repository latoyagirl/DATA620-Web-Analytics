# HW1 -  Hello_Graph World

Import all libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Create an empty graph

```python
G=nx.Graph()
```

Create nodes

```python
nodes = ['Andre', 'Beverly', 'Carol', 'Diane', 
        'Ed', 'Fernando', 'Heather', 'Ike', 'Jane']

G.add_nodes_from(nodes)
```



Explicitly define position of each node



```python
pos={'Andre':(0,1),
     'Beverly':(0,-1),
     'Carol':(1,2),
     'Diane':(1,0),
     'Ed':(1,-2),
     'Ike':(4,0),
     'Garth':(2,-1),
     'Jane':(5,0),
     'Fernando':(2,1),
     'Heather':(3,0)}
```

Add edges

```python
edges = [('Andre', 'Beverly'),
         ('Andre', 'Carol'),
         ('Andre', 'Diane'),
         ('Andre', 'Fernando'),
         ('Beverly', 'Diane'),
         ('Beverly', 'Garth'),
         ('Beverly', 'Ed'),
         ('Carol', 'Fernando'),
         ('Carol', 'Diane'),
         ('Ed', 'Diane'),
         ('Ed', 'Garth'),
         ('Diane', 'Fernando'),
         ('Diane', 'Garth'),
         ('Fernando','Garth'),
         ('Heather', 'Fernando'),
         ('Heather', 'Garth'),
         ('Heather', 'Ike'),
         ('Ike','Jane' )
        ]
G.add_edges_from(edges)
```

- - -

```python

nx.draw(G,pos, with_labels=True,font_weight="bold",node_color='skyblue', node_size=1500)
plt.show()
```
